```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data = {'Pre-Trained Models':['VGG-16', 'Inception-V3', 'ResNet50', 'MobileNetV2'],
        'Score':[99, 98, 60, 90]}
  
df = pd.DataFrame(data, index =['rank1',
                                'rank2',
                                'rank3',
                                'rank4'])
  

df.to_csv('graphs.csv', index = False, header = True)
data = pd.read_csv('graphs.csv')
data = pd.read_csv('graphs.csv')
```

```python
sns.set_theme(style="whitegrid")
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set(font_scale=2)
ax = sns.barplot(x="Pre-Trained Models", y="Score", data=data)
```

```python
plt.figure(figsize=(18, 12))
splot=sns.barplot(x="Pre-Trained Models",y="Score",data=data)
for p in splot.patches:
    splot.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')
plt.xlabel("Pre-Trained Models", size=14)
plt.ylabel("LifeExp", size=14)
```

```python
X = ['VGG-16','Inception-V3','ResNet50','MobileNetV2']
precision = [98,20,20,40]
f1 = [20,30,25,30]
recall= [20,30,25,30]
X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, precision, 0.2, label = 'Precision')
plt.bar(X_axis + 0.2, f1, 0.2, label = 'F-1')
plt.bar(X_axis + 0.0, recall, 0.2, label = 'Recall')
  
plt.xticks(X_axis, X)
plt.xlabel("Pre-Trained Models")
plt.ylabel("Number of Students")
plt.title("Number of Students in each group")
plt.legend()
plt.show()
```
